```python
from google.cloud import translate

source = 'en'
target = 'id'
translate_client = translate.Client()

def translate_caption(caption):
    translation = translate_client.translate(
    caption,
    target_language=target,
    source_language=source)
    
    return translation['translatedText']
```

```python
indo_dataset = []
file_indo = open('Flickr8k.token.indo29048.txt', 'w')
file = open('Flickr8k.token.txt', 'r')

for line in file.read().split('\n')[29048:]:
    line = line.split('\t')
    caption = line[-1]    
    indo_caption = translate_caption(caption)

    line[-1] = indo_caption
    data = '\t'.join(line)
    
    indo_dataset.append(data)

for cap in indo_dataset:
    file_indo.write(cap + '\n')

print(indo_dataset)
file_indo.close()
file.close()
```
